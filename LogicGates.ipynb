```python
from tabulate import tabulate


def AND_gate_custom():
    input1 = [0, 0, 1, 1]
    input2 = [0, 1, 0, 1]
    weight1 = weight2 = [1, 1, 1, 1]
    threshold = 2
    result = []
    for index in range(len(input1)):
        output = 1 if (input1[index] * weight1[index] + input2[index] * weight2[index]) >= threshold else 0
        result.append([input1[index], input2[index], weight1[index], weight2[index], threshold, output])
    print("AND Gate Simulation")
    print(tabulate(result, headers=["Input1", "Input2", "Weight1", "Weight2", "Threshold", "Output"]))

def OR_gate_custom():
    input1 = [0, 0, 1, 1]
    input2 = [0, 1, 0, 1]
    weight1 = weight2 = [1, 1, 1, 1]
    threshold = 1
    result = []
    for index in range(len(input1)):
        output = 1 if (input1[index] * weight1[index] + input2[index] * weight2[index]) >= threshold else 0
        result.append([input1[index], input2[index], weight1[index], weight2[index], threshold, output])
    print("OR Gate Simulation")
    print(tabulate(result, headers=["Input1", "Input2", "Weight1", "Weight2", "Threshold", "Output"]))

def NOT_gate_custom():
    input_val = [0, 1]
    weight = [-1, -1]
    threshold = 0
    result = []
    for index in range(len(input_val)):
        output = 1 if (input_val[index] * weight[index]) >= threshold else 0
        result.append([input_val[index], weight[index], threshold, output])
    print("NOT Gate Simulation")
    print(tabulate(result, headers=["Input", "Weight", "Threshold", "Output"]))

def NAND_gate_custom():
    input1 = [0, 0, 1, 1]
    input2 = [0, 1, 0, 1]
    weight1 = weight2 = [-1, -1, -1, -1]
    threshold = -2
    result = []
    for index in range(len(input1)):
        output = 1 if (input1[index] * weight1[index] + input2[index] * weight2[index]) > threshold else 0
        result.append([input1[index], input2[index], weight1[index], weight2[index], threshold, output])
    print("NAND Gate Simulation")
    print(tabulate(result, headers=["Input1", "Input2", "Weight1", "Weight2", "Threshold", "Output"]))

def NOR_gate_custom():
    input1 = [0, 0, 1, 1]
    input2 = [0, 1, 0, 1]
    weight1 = weight2 = [1, 1, 1, 1]
    threshold = 0
    result = []
    for index in range(len(input1)):
        output = 1 if (input1[index] * weight1[index] + input2[index] * weight2[index]) <= threshold else 0
        result.append([input1[index], input2[index], weight1[index], weight2[index], threshold, output])
    print("NOR Gate Simulation")
    print(tabulate(result, headers=["Input1", "Input2", "Weight1", "Weight2", "Threshold", "Output"]))

def XOR_gate_custom():
    input1 = [0, 0, 1, 1]
    input2 = [0, 1, 0, 1]
    weight1 = weight2 = weight3 = weight6 = [1, 1, 1, 1]
    weight4 = weight5 = [-1, -1, -1, -1]
    threshold1 = [0.5, 0.5, 0.5, 0.5]
    threshold2 = [-1.5, -1.5, -1.5, -1.5]
    threshold3 = [1.5, 1.5, 1.5, 1.5]

    def XOR_func(a, b):
        return 1 if a != b else 0

    result = []
    for index in range(len(input1)):
        output = XOR_func(input1[index], input2[index])
        result.append([input1[index], input2[index], weight1[index], weight2[index], weight3[index], weight4[index], weight5[index], weight6[index], threshold1[index], threshold2[index], threshold3[index], output])

    print("XOR Gate Simulation")
    print(tabulate(result, headers=["Input1", "Input2", "Weight1", "Weight2", "Weight3", "Weight4", "Weight5", "Weight6", "Threshold1", "Threshold2", "Threshold3", "Output"]))


AND_gate_custom()
print("\n")
OR_gate_custom()
print("\n")
NOT_gate_custom()
print("\n")
NAND_gate_custom()
print("\n")
NOR_gate_custom()
print("\n")
XOR_gate_custom()
```
